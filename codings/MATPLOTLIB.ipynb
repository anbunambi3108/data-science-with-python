```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,11,10)
y=x**2
```

```python
plt.plot(x,y,'red',marker='<')
plt.xlabel("sample")
plt.ylabel("py")
plt.title('graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'red')
plt.subplot(1,2,2)
plt.plot(y,x,'black')
```
