```python
import matplotlib.pyplot as plt
import numpy  as np
%matplotlib inline
np.random.seed(5)
```

```python
x=np.arange(1,20,2)
y=x**1.3+np.random.rand(*x.shape)*20.0
s=np.random.rand(*x.shape)*500+500 #to create size of markers without using size function
plt.scatter(x,y,s,c='g',alpha=1,marker=r'$\clubsuit$',label='luck')
plt.xlabel('thu')
plt.ylabel('happy')
plt.legend(loc='best')
```
