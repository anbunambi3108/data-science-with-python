```python
#THREE SHEETS IN ONE EXCEL FILE:
import pandas as pd
d1={
    'v':['stefan','ian','paul','damon'],
     'm1':[4,2,3,5]}
d2={
    'h':['jermy','elena','nina','matt'],
    'm2':[3,5,3,4]}
d3={
    'w':['bonnie','cal','ben','mick'],
    'm3':[1,3,5,2]}
data1=pd.DataFrame(d1)
data2=pd.DataFrame(d2)
data3=pd.DataFrame(d3)

with pd.ExcelWrite('data.xlsx') as w:
    data1.to_excel(w,sheet_name='v')
    data2.to_excel(w,sheet_name='h')
    data3.to_excel(w,sheet_name='m')
    
```
