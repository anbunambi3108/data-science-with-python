# object oriented method:

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.2,.8,.9])
axes2=fig.add_axes([.2,.6,.4,.4])
```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.2,.8,.9])
axes2=fig.add_axes([.2,.6,.4,.4])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.2,.8,.9])
axes2=fig.add_axes([.2,.6,.4,.4])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_xlabel('num')
axes1.set_ylabel('sqr')
axes2.set_ylabel('num')
axes2.set_xlabel('sqr')
```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.2,.8,.9])
axes2=fig.add_axes([.2,.6,.4,.4])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_xlabel('num')
axes1.set_ylabel('sqr')
axes2.set_ylabel('num')
axes2.set_xlabel('sqr')
axes1.set_title('sample')
axes2.set_title('sample1')
```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.2,.8,.9])
axes1.plot(x,label='num')
axes1.plot(y,label='sqr')
plt.legend(loc='best')
```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.2,.8,.9])
axes1.plot(x,label='num',color='lightblue',lw=2,ls='--',marker='*',markersize='10',markerfacecolor='w',markeredgecolor='k',markeredgewidth='.2')
axes1.plot(y,label='sqr',color='black',lw=2,ls=':',marker='*',markersize='10',markerfacecolor='w',markeredgecolor='k',markeredgewidth='.5')
plt.legend(loc='best')
```
