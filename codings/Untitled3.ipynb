```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
x=[[182,72],[170,56],[168,60],[179,68],[182,72],[188,77]]
x1=np.array(x)
plt.scatter(x1[:,0],x1[:,1],s=50)
```

```python
KMeans=KMeans(n_clusters=2)
KMeans.fit(x1)
yKMean=KMeans.predict(x1)
```

```python
plt.scatter(x1[:,0],x1[:,1],cmap='coolwarm',c=yKMean,s=50)
center=KMeans.cluster_centers_
plt.scatter(center[:,0],center[:,1],c='k',s=150)
print(center)
```
