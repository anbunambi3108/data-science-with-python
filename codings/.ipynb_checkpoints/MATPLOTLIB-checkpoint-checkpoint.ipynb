```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,11,10)
y=x**2
```

```python
plt.plot(x,y,'red',marker='<')
plt.xlabel("sample")
plt.ylabel("py")
plt.title('graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'red')
plt.subplot(1,2,2)
plt.plot(y,x,'black')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x=[1,2,3,4,5]
y=[9,8,7,6,3]
plt.plot(x,y,'m',marker='*')
plt.xlabel('num1')
plt.ylabel('num2')
plt.title('sample')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'r')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r',marker='|')
plt.xlabel('s1')
plt.ylabel('s2')
plt.title('samp')
plt.subplot(2,2,4)
plt.plot(x,y,'k',marker='D')
plt.xlabel('a1')
plt.ylabel('a2')
plt.title('samp2')
plt.tight_layout()
```

```python
fig=plt.figure()
axes=fig.add_axes([2,3,4,5])
axes.plot(x,y)
axes.set_xlabel('s1')
axes.set_ylabel('s2')
axes.set_title('samp')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.2,0.3,0.4,0.5])
axes.plot(x,y)
axes.set_xlabel('s1')
axes.set_ylabel('s2')
axes.set_title('samp')
```
