```python
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('weather.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df['Temperature '])
```

```python
sns.distplot(df['Wind Speed'])
```

```python
sns.jointplot(df['Wind Speed'],df['Temperature '],kind='kde',color='blue')
```

```python
sns.jointplot(df['Wind Speed'],df['Temperature '],kind='reg',color='m')
```

```python
sns.jointplot(df['Wind Speed'],df['Temperature '],kind='hex',color='m')
```

```python
sns.jointplot(df['Wind Speed'],df['Temperature '],kind='scatter',color='m')
```

```python
sns.rugplot(df['Wind Speed'],axis='x',color='b',height=0.3)
sns.rugplot(df['Temperature '],axis='y',color='k',height=0.3)
```

```python
sns.countplot(x='Event',data=df)
```

```python
sns.barplot(x='Wind Speed',y='Temperature ',data=df)
```

```python
sns.violinplot(x='Temperature ',y='Event',data=df)
```

```python
sns.boxplot(x='Temperature ',y='Wind Speed',data=df,hue='Event')
```

```python
import numpy as np
sns.barplot(x='Wind Speed',y='Temperature ',data=df,estimator=np.std)
```

```python
sns.countplot(x='Wind Speed',data=df)
```

```python
sns.barplot(x='Event',y='Temperature ',data=df)
```
